```python
import random, math
import pandas as pd
import tsplib95 as tl
import numpy as np
import string
```

```python
def save_instance(instance_name,instances):
    with open(F"../data/random/{instance_name}.txt", "w") as f:
        for values in instances:
            print(values,file=f)       

def save_index(dataset, instances):
    with open(F"instancias/{dataset}/instances.txt", "w") as f:
        for instance in instances: 
            print(instance, file=f)
```

# Generacion de datos

## Sin Repetidos

```python

cantidad_de_archivos = 10
tamaño_archivo = 5
contenedor = []
for i in range(1,cantidad_de_archivos): #cantidad de archivos
    archivo = []
    for j in range(0,tamaño_archivo):
        archivo.append(''.join(random.choice(string.ascii_letters + string.digits) for _ in range(i)))
    contenedor.append(archivo)
    save_instance("sinRepetidos"+F"-{tamaño_archivo}", archivo)
    print(archivo)
    tamaño_archivo += tamaño_archivo 
#print(contenedor)
```

## Con Repetidos

```python

cantidad_de_archivos = 10
tamaño_archivo = 5
contenedor = []
repeticiones = np.random.randint(0,tamaño_archivo*cantidad_de_archivos, size=tamaño_archivo+1)
print(repeticiones)
for i in range(1,cantidad_de_archivos): #cantidad de archivos
    archivo = []
    for j in range(0,tamaño_archivo):
        currentWord = ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(i))
        print(repeticiones[j])
        for k in range(0,repeticiones[j]):
            archivo.append(currentWord)
    contenedor.append(archivo)
    save_instance("conRepetidos"+F"-{tamaño_archivo}", archivo)
    print(archivo)
    tamaño_archivo += tamaño_archivo 
```
